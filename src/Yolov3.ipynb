```
import os
import glob
import time

img_dir = "./data/images/"
data_path = os.path.join(img_dir, '*g')
files = glob.glob(data_path)
f1 = open("train.txt", "w")
f2 = open("test.txt", "w")
for dirpath, dirs, files in os.walk(img_dir):  
  print('1. %s ' % dirpath)
  print('2. %s' % dirs)
  print('3. %s' % files)
  arr = ['set02', 'set05', 'set08']
  for filename in files: 
    fname = os.path.join(dirpath,filename) 
    #print('4. %s' % filename)
    if fname.endswith('.jpg') and fname.find('visible') != -1:
        #time.sleep(1)
        print('5. %s' %fname)
        if any(f in fname for f in arr):
            print('train')
            f1.write(fname+'\n')
        else:
            print('test')
            f2.write(fname+'\n')
              
```
