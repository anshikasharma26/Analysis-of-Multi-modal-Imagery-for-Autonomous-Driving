```
#!rm -f train.txt test.txt
import os
import glob
import time
import cv2

img_dir = "./data/images/"
data_path = os.path.join(img_dir, '*g')
files = glob.glob(data_path)

f1 = open("train.txt", "w")
f2 = open("test.txt", "w")
for dirpath, dirs, files in os.walk(img_dir):  
  for filename in files: 
    fname = os.path.join(dirpath,filename) 
    if fname.endswith('.jpg'): 
      print('5. %s' %fname)
      f1.write(fname+'\n')
      f2.write(fname+'\n')
f1.close()
f2.close()
```
